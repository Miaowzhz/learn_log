```python
# 成绩判断
#
# 需求：输入分数，输出对应等级
# 思路：1) 获取分数输入 2) 使用if-elif-else判断等级 3) 输出等级结果
num = int(input("请输入你的分数: "))
if num >= 90:
	print("你的成绩等级是A")
elif num >= 70:
	print("你的成绩等级是B")
elif num >= 50:
	print("你的成绩等级是C")
else:
	print("你的成绩等级是D")
```

```python
# 乘法表
#
# 需求：打印九九乘法表
# 思路：1) 外层循环控制行 2) 内层循环控制列 3) 格式化输出
for i in range(1, 10):
	for j in range(1, i+1):
		print(f"{j} x {i} = {j * i}", end = "\t")
	print("\n")
```

```python
import random

# 猜数字游戏
#
# 需求：实现猜数字游戏
# 思路：1) 生成随机数 2) 循环获取用户猜测 3) 给出提示直到猜中
random_num = random.randint(1,100)
print(random_num)
while True:
	input_num = int(input("请输入所猜测的数字: "))
	if input_num == random_num:
		print(f"你猜对了!随机数是{random_num}")
		break
	elif input_num > random_num:
		print("你猜大了! 请继续")
	else:
		print("你猜小了! 请继续")
```
