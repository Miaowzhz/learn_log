```python
# 1.2 if条件判断进阶
# 判断是否是闰年
year = int(input("请输入年份: "))

if year % 100 != 0 and year % 4 == 0:
    print(f"{year} 是闰年")
elif year % 400 == 0:
    print(f"{year} 是闰年")
else:
    print(f"{year} 不是闰年")

if (year % 100 != 0 and year % 4 == 0) or (year % 400 == 0):
    print(f"{year} 是闰年")
else:
    print(f"{year} 不是闰年")
```

```python
# 1.5 案例 - 三角形类型判断
# pass 空语句, 起到语法站位的作用
a = int(input("请输入边长a: "))
b = int(input("请输入边长b: "))
c = int(input("请输入边长c: "))

is_true = a + b > c and a + c > b and b + c > a

if is_true:
    if a == b == c:
        print("等边三角形")
    elif a == b or a == c or b == c:
        print("等腰三角形")
    else:
        print("普通三角形")
else:
    print("不是三角形")
```

```python
# 3.1 匹配模式

day = int(input("请输入星期几(1-7): "))
match day:
    case 1:
        print("星期一")
    case 2:
        print("星期二")
    case 3:
        print("星期三")
    case 4:
        print("星期四")
    case 5:
        print("星期五")
    case 6 | 7:
        print("周末休息")
    case _:
        print("输入错误")
```

```python
# 3.2-练习
a = float(input("请输入数字a: "))
_char = input("请输入运算符号(+ - * /): ")
b = float(input("请输入数字b: "))

match _char:
    case "+":
        print(a + b)
    case "-":
        print(a - b)
    case "*":
        print(a * b)
    case "/" if b != 0:
        print(a / b)
    case _:
        print("输入错误")
```

```python
# 4.1 while 循环
i = 1
while i <= 10:
    print(f"Python: {i}")
    i += 1
else:
    print("循环结束")
```

```python
# 4.2 案例

num = 1
num_sum = 0
while num <= 100:
    if num % 2 == 0:
        num_sum += num
    num += 1

print(f"1-100之间所有偶数的累加之和: {num_sum}")
```

```python
# 4.3 for循环 - 轮询遍历机制
# 字符串

msg = input("请输入需要遍历的字符串")
for s in msg:
    print(s)
else:
    print("循环结束")
```

```python
# 4.4 for循环 案例
total = 0
for i in range(1, 101):
    if i % 2 == 1:
        total += i
    i += i

print(f"1-100之间所有奇数的累加之和: {total}")


total1 = 0
for i in range(100, 501):
    if i % 3 == 0:
        total1 += i
    i += i

print(f"100-500之间所有3的倍数的数字之和: {total1}")
```

```python
# 4.5 嵌套循环
for i in range(1, 10):
    for j in range(1, i + 1):
        print(f"{j} x {i} = {i * j}", end="\t")
    print()
```
