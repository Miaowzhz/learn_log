```python
# 个人信息展示
# 需求：用变量存储姓名和年龄，然后输出
# 思路：创建变量 → 赋值 → 打印，尝试用不同方式拼接字符串
name: str = "wmy"
age: int = 24
print(name)
print(age)
print("姓名：", name, "年龄：", age)
```

```python
# 变量交换
# 需求：交换两个变量的值并输出交换前后的结果
# 思路：1) 创建两个变量赋值 2) 使用临时变量或Python特有方式交换 3) 打印验证
name = "wmy"
age = 24
print(name)
print(age)

name, age = age, name
print(name)
print(age)
```

```python
# 简单计算器
# 需求：输入两个数字，输出四则运算结果
# 思路：1) 使用input获取两个数字 2) 进行类型转换 3) 计算并输出结果
a = input("请输入第一个数字：")
b = input("请输入第二个数字：")
a = int(a)
b = int(b)
print(a + b)
print(a - b)
print(a * b)
print(a / b)
```
