```python
# 常见的数据类型
"""
int float str bool NoneType
"""
print(type(3.14))
name = "mark"
print(type(name))

print(type(None))

# isinstance 判断数据是否是指定类型,返回布尔类型
print(isinstance(name, str))
print(isinstance(name, int))
```

```python
# 字符串
# 定义的三种方式
s1 = "hello"
s2 = 'hello'
s3 = """
    你好:
        哈喽
        那你
"""

print(s3)
print(isinstance(s3, str))
print(isinstance(s3, int))

# 转义字符
s2 = "It's very interesting"
print(s2)
s2 = 'It\'s very interesting'

msg = "Hello 的意思是 \"您好\""
print(msg)
print("您好! \n\t 一键三连")
```

```python
# 字符串拼接
name = "mark"
age = 18
_class = "区块链工程"
_hobby = "Python"

# 将int类型的数字转成字符串: str()
print("欢迎! " + name + " 的到来")
print("大家好! 我是" + name + ", 今年" + str(age) + "岁, 学习的是" + _class + ", 爱好是" + _hobby)

# 加号拼接比较繁琐 破坏字符串的完整性 需要类型转换
# 字符串的格式化方式
# 方式一: %s
print("大家好, 我是 %s , 今年 %s 岁" % (name, age))

# 方式二: format
print("大家好, 我是 {} , 今年 {} 岁".format(name, age))

# 方式三: f-string 主流方式
# print(f"大家好, 我是 {name} , 今年 {age} 岁")
print(f"大家好! {name}, 今年 {age} 岁")
```

```python
# 运算符
# 算数运算符 + - * / // % **
# 算数运算符的优先级 优先级从高到低
# ** 大于 * / // % 大于 + -

a = float(input("请输入第一个数: "))
b = float(input("请输入第二个数: "))

print(f"a + b = {a + b}, a - b = {a - b}")

# 浮点数进行运算的时候会出现精度丢失
# 计算机底层是二进制, 浮点数无法精确表示


```

```python
# 比较运算符 > < >= <= == !=
# 计算运算符两边的表达式,返回布尔类型结果
print("100 == 100 吗 : ", 100 == 100)
print("'100' == '100' 吗 : ", '100' == '100')
print("100 != 100 吗 : ", 100 != 100)

```

```python
# 逻辑运算符 and or not
num = int(input("请输入一个数字: "))
print(f"{num} 是否在10-20之间: ", num >= 10 and num <= 20)
print(f"{num} 是否不在10-20之间: ", num < 10 or num > 20)
```
