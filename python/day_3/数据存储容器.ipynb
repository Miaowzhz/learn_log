```python
# 练习 1: list 列表 - 创建、索引、修改
# 要求:
# 1. 创建一个列表 names, 保存 3 个同学的名字
# 2. 打印第 1 个同学
# 3. 打印最后 1 个同学
# 4. 把第 2 个同学改成一个新名字
# 5. 打印修改后的列表

# 在下面写代码
names = list()
names = ["mark", "alice", "tom"]
print(names[0])
print(names[-1])

names[1] = "luna"
for name in names:
	print(name)
```

```python
# 练习 2: list 列表 - 增加、删除、遍历
# 要求:
# 1. 创建一个空列表 tasks
# 2. 使用 append 添加 3 个待办事项
# 3. 删除其中 1 个待办事项
# 4. 使用 for 循环逐个打印剩余事项

# 在下面写代码
tasks = []
tasks.append("复习")
tasks.append("吃饭")
tasks.append("学习")
tasks.remove("吃饭")
tasks.pop()
for task in tasks:
	print(task)
```

```python
# 练习 3: list 切片
# 已知 nums = [10, 20, 30, 40, 50, 60]
# 要求:
# 1. 取出前三个元素
# 2. 取出从索引 2 到最后的元素
# 3. 取出索引 1 到索引 4 之间的元素
# 4. 反转整个列表

nums = [10, 20, 30, 40, 50, 60]

# 在下面写代码
nums_1 = nums[:3]
print(nums_1)

nums_2 = nums[2:]
print(nums_2)

nums_3 = nums[1:4]
print(nums_3)

nums_4 = nums[::-1]
print(nums_4)
```

```python
# 练习 5: tuple 元组 - 创建、索引、解包
# 要求:
# 1. 创建一个元组 student, 保存姓名、年龄、成绩
# 2. 使用索引打印姓名
# 3. 使用解包把姓名、年龄、成绩分别赋值给 name、age、score
# 4. 使用 f-string 输出: 姓名: xxx, 年龄: xxx, 成绩: xxx

# 在下面写代码
student = ("mark", 18, 90)

print(student[0])

name, age, score = student

print(f"姓名: {name}, 年龄: {age}, 成绩: {score}")

for i in student:
	print(i)
```

```python
# 练习 6: set 集合 - 去重、添加、删除
# 已知 nums = [1, 2, 2, 3, 3, 3, 4]
# 要求:
# 1. 把 nums 转成集合, 完成去重
# 2. 添加数字 5
# 3. 删除数字 1
# 4. 打印最终集合

nums = [1, 2, 2, 3, 3, 3, 4]

# 在下面写代码
nums_set = set()
print(type(nums_set))
for num in nums:
	nums_set.add(num)
print(nums_set)

nums_set.add(5)
print(nums_set)

nums_set.remove(1)
print(nums_set)

for num in nums_set:
	print(num)
```

```python
# 练习 7: dict 字典 - 增删改查
# 要求:
# 1. 创建一个 student 字典, 包含 name、age、score 三个 key
# 2. 打印 name 对应的值
# 3. 把 score 修改为 100
# 4. 新增 hobby, 值自定
# 5. 删除 age
# 6. 打印最终字典

# 在下面写代码
student = {"name": "mark", "age": 18, "score": 90}

print(student["name"])

student["score"] = 100

student["hobby"] = "Python"


del student["age"]

print(student)
```

```python
# 综合练习: 学生信息列表
# 要求:
# 1. 创建 students 列表, 里面保存 3 个学生字典
#    每个学生字典包含 name、age、score
# 2. 使用 for 循环打印每个学生的信息
# 3. 只打印成绩 >= 80 的学生姓名
# 4. 计算所有学生的平均分

# 示例结构:
# students = [
#     {"name": "张三", "age": 18, "score": 90},
#     {"name": "李四", "age": 19, "score": 75},
# ]

# 在下面写代码
student_1 = {"name": "mark", "age": 19, "score": 90}
student_2 = {"name": "ailice", "age": 19, "score": 87}
student_3 = {"name": "lucy", "age": 19, "score": 76}

students = list()

students.append(student_1)
students.append(student_2)
students.append(student_3)

score_sum = 0
total = 0

for student in students:
	score_sum += student["score"]
	total += 1
	print(student)

for student in students:
	if student["score"] >= 80:
		print(student["name"])

print(score_sum)
print(total)
score_v = score_sum / total
print(score_v)
```
