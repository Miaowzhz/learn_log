```python
# 待办事项管理
#
# 需求：使用列表实现待办事项的增删查
# 思路：1) 创建空列表 2) 实现添加、删除、查看功能 3) 使用循环展示菜单
tasks = list()

while True:
	print("------------------------------")
	print("1: 添加, 2: 删除, 3: 查看 4: 退出")
	print("------------------------------")
	flat = input("请输入操作: ")

	match flat:
		case "1":
			task = input("请输入要添加的待办事项: ")
			tasks.append(task)
			print("添加成功")
		case "2":
			task = input("请输入要删除的待办事项: ")
			tasks.remove(task)
			print("删除成功")
		case "3":
			print(f"所有待办事项: {tasks}")
		case "4":
			print("已退出")
			break
		case _:
			print("请输入正确的操作")
```

```python
# 字符串处理
#
# 需求：实现文本统计功能
# 思路：1) 输入一段文本 2) 统计字符数、单词数、行数 3) 查找替换功能

msg = input("请输入一段文本: ")

while True:
	print("-------------------------------------------")
	print("1: 统计, 2: 查找, 3: 替换, 4: 查看文本, 5: 退出")
	print("-------------------------------------------")
	choice = input("请输入操作: ")

	if choice == "1":
		char_count = len(msg)
		word_count = len(msg.split())
		line_count = msg.count("\n") + 1 if msg else 0

		print(f"字符数: {char_count}")
		print(f"单词数: {word_count}")
		print(f"行数: {line_count}")

	elif choice == "2":
		keyword = input("请输入需要查找的内容: ")
		count = msg.count(keyword)

		if count > 0:
			print(f"已找到 {keyword}, 一共出现 {count} 次")
		else:
			print(f"没有找到: {keyword}")

	elif choice == "3":
		old_word = input("请输入想要替换的内容: ")
		new_word = input("请输入替换后的内容: ")

		if old_word in msg:
			msg = msg.replace(old_word, new_word)
			print(f"替换成功, 当前文本: {msg}")
		else:
			print(f"没有找到需要替换的内容: {old_word}")

	elif choice == "4":
		print(f"当前文本: {msg}")

	elif choice == "5":
		print("已退出")
		break

	else:
		print("请输入正确的操作")
```

```python
# 学生信息管理系统
#
# 需求：使用字典存储学生信息
# 思路：1) 创建字典存储学生信息 2) 实现增删改查功能 3) 使用循环展示菜单

students = dict()
while True:
	print("--------------------------------------")
	print("1: 添加, 2: 删除, 3: 修改 4: 查看 5: 退出")
	print("--------------------------------------")
	flat = input("请输入操作: ")

	match flat:
		case "1":
			name = input("请输入需要添加的学生姓名: ")
			age = int(input("请输入需要添加的学生年龄: "))
			score = float(input("请输入需要添加的学生成绩: "))

			if name in students.keys():
				print("该学生的信息已存在!")
				continue
			else:
				students[name] = {"age": age, "score": score}
				print("添加成功")
		case "2":
			name = input("请输入需要删除的学生姓名: ")
			if name not in students.keys():
				print("该学生的信息不存在!")
				continue
			else:
				del students[name]
				print("删除成功")
		case "3":
			name = input("请输入需要修改的学生姓名: ")
			age = int(input("请输入新的学生年龄: "))
			score = float(input("请输入新的学生成绩: "))

			if name not in students.keys():
				print("该学生的信息不存在!")
				continue
			else:
				students[name] = {"age": age, "score": score}
				print("修改成功")
		case "4":
			for name in students.keys():
				student = students[name]
				print(f"姓名: {name}, 年龄: {student['age']}, 成绩: {student['score']}")
		case "5":
			print("已退出")
			break
		case _:
			print("请输入正确的操作")
```
