```python
# 函数练习
#
# 需求：实现各种实用函数
# 思路：1) 实现计算函数（加减乘除）2) 实现字符串处理函数 3) 使用lambda简化代码

# 实现计算函数
def ji_suan(a,b,op):
	match op:
		case '+':
			return a+b
		case '-':
			return a-b
		case '*':
			return a*b
		case '/':
			if b == 0:
				return "0不能做除数"
			return a/b
		case _:
			return "参数不合法"

print(ji_suan(1,2,'+'))
print(ji_suan(1,2,'/'))
print(ji_suan(1,0,'/'))
print(ji_suan(1,2,'*'))
print(ji_suan(1,2,'-'))

# 实现字符串处理函数
def string_handle(s):
    return "是回文数" if s == s[::-1] else "不是回文数"
print(string_handle("12321"))

# 使用lambda简化代码
string_handle = lambda s: "是回文数" if s == s[::-1] else "不是回文数"
print(string_handle("12321"))
```

```python
# 递归练习
#
# 需求：使用递归解决问题
# 思路：1) 实现阶乘计算 2) 实现斐波那契数列 3) 理解递归的执行过程
# 实现阶乘计算
def jie_cheng(n):
	if n == 0:
		return 1

	return n * jie_cheng(n-1)
print(jie_cheng(5))

# 实现斐波那契数列
def fei_bo_na_qi(n):
    if n == 0:
        return 0
    elif n == 1:
        return 1
    else:
        return fei_bo_na_qi(n-1) + fei_bo_na_qi(n-2)
```
