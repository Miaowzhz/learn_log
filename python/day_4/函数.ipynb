## 第一阶段：函数基础

函数就是一段被命名的代码。把重复逻辑放进函数后，需要时通过函数名调用。

- 参数：调用函数时交给函数处理的数据
- 返回值：函数处理完成后交回来的结果
- `print()`：把内容显示在屏幕上
- `return`：把结果返回给调用函数的位置，并结束本次函数调用

```python
# 1.2 参数与返回值
def add(num1, num2):
    result = num1 + num2
    return result

total = add(10, 20)
print(total)
```

```python
# 1.3 函数说明文档：写在函数内部的第一行
def get_level(score):
    """根据分数返回成绩等级。"""
    if score >= 90:
        return "优秀"
    elif score >= 60:
        return "及格"
    return "不及格"

print(get_level(85))
help(get_level)
```

```python
# 1.4 函数嵌套调用：一个函数内部调用另一个函数
def get_total(prices):
    return sum(prices)

def print_bill(prices):
    total = get_total(prices)
    print(f"商品数量: {len(prices)}")
    print(f"总价: {total}")

print_bill([10, 20, 30])
```

```python
# 练习4：容器与函数
# 定义 count_pass(scores)，接收分数列表，返回及格人数（分数 >= 60）
# 使用 [58, 90, 76, 45, 60] 测试

# 在这里写代码
def count_pass(scores):
    count = 0

    for score in scores:
        if score >= 60:
            count += 1

    return count
print(count_pass([58, 90, 76, 45, 60]))
```

```python
# 练习5：函数嵌套调用
# 定义 get_average(scores)，返回平均分
# 定义 print_report(name, scores)，在内部调用 get_average
# 最终输出示例：小明的平均分是: 80.0
# 使用名字"小明"和分数 [70, 80, 90] 测试

# 在这里写代码
def get_average(scores):
	total = 0
	count = len(scores)
	for score in scores:
		total += score
	return total/count

def print_report(name, scores):
	average = get_average(scores)
	print(f"{name}的平均分是: {average}")

print_report("小明", [70, 80, 90])
```

```python
def triangle_area(base, height):
    return base * height / 2

print(triangle_area(10, 5))

def sum(s_list):
    s_sum = 0
    for s in s_list:
        s_sum += s
    return s_sum

def score(stu_list):
    stu_count = len(stu_list)
    return max(stu_list), min(stu_list), round(sum(stu_list) / stu_count, 1)

print(score([90, 80, 70, 60, 50]))
```

## 第二阶段：作用域与参数进阶

### 2.1 变量作用域

- 全局变量：定义在函数外部，当前文件的多个函数可以读取
- 局部变量：定义在函数内部，只在本次函数调用期间有效
- 函数内部优先使用局部变量；局部变量与全局变量同名时，它们仍是两个变量

```python
# 2.1 作用域示例
name = "全局小明"

def show_name():
    name = "局部小红"
    print(f"函数内部: {name}")

show_name()
print(f"函数外部: {name}")
```

### 2.2 参数传递

调用函数时，形参会指向传入的对象。

- 在函数内给形参重新赋值，不会改变外面的变量
- 在函数内直接修改传入的列表，外面的列表也会发生变化
- 如果不想修改原列表，可以传入切片副本：`function_name(nums[:])`

```python
# 2.2 参数传递示例
def change_number(number):
    number = 100

def add_score(scores):
    scores.append(100)

num = 10
score_list = [60, 80]
change_number(num)
add_score(score_list)
print(num)
print(score_list)
```

### 2.3 默认参数、关键字参数和不定长参数

- 默认参数：调用时不传值，就使用预先设置的值
- 关键字参数：通过参数名传值，顺序更灵活
- `*args`：接收任意数量的位置参数，在函数内是元组
- `**kwargs`：接收任意数量的关键字参数，在函数内是字典

默认参数必须放在普通参数后面。初学阶段不要把列表或字典直接作为默认值。

```python
# 2.3 参数进阶示例
def introduce(name, city="北京"):
    print(f"我叫{name}，来自{city}")

introduce("小明")
introduce(city="上海", name="小红")

def add_all(*numbers):
    return sum(numbers)

def show_student(**student):
    print(student)

print(add_all(10, 20, 30))
show_student(name="小明", age=18, score=90)
```

```python
# 练习7：列表参数
# 定义 remove_fail(scores)，直接删除列表中所有低于 60 的分数
# 使用 scores = [45, 80, 59, 60, 90] 测试，并在调用后打印原列表
# 提示：遍历列表副本 scores[:]，从原列表 scores 中删除元素

# 在这里写代码

def remove_fail(scores):
	for score in scores:
		if score < 60:
			scores.remove(score)

	return scores

print(remove_fail([45, 80, 59, 60, 90]))
```

```python
# 练习8：默认参数与关键字参数
# 定义 create_user(name, role="普通用户")，返回包含 name 和 role 的字典
# 第一次只传入名字"小明"
# 第二次使用关键字参数传入 role="管理员"、name="小红"，故意交换参数顺序

# 在这里写代码
def create_user(name, role="普通用户"):
	return {"name": name, "role": role}
print(create_user("小明"))
print(create_user(role="管理员", name="小红"))
```

```python
# 练习9：*args
# 定义 get_max(*numbers)，返回所有参数中的最大值
# 不使用 max()，通过循环完成
# 使用 get_max(35, 90, 12, 78) 测试，结果应为 90

# 在这里写代码
def get_max(*numbers):
	max_num = 0
	for num in numbers:
		if num > max_num:
			max_num = num

	return max_num
print(get_max(35, 90, 12, 78))
```

```python
# 练习10：**kwargs
# 定义 print_info(**info)，遍历并打印每个 key 和 value
# 使用 print_info(name="小明", age=18, city="北京") 测试

# 在这里写代码
def print_info(**info):
    for key, value in info.items():
        print(f"{key}: {value}")

print_info(name="小明", age=18, city="北京")
```
